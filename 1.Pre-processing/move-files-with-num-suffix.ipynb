```python
import glob
import shutil
import os

#counter
#see: https://pynative.com/python-count-number-of-files-in-a-directory/

#parent directory which contains the folders that ocntains the subfiles to combine:
source_parent_folder=r"C:/Users/Z6/Desktop/3DSEM//Stefan/1.AT/2.June20_v4_final-fr-with-vess2-small-408R/5.ALIGNED/Batch-alignment1/0.source-images"


# folder path to one of the subfolder which contains vessels for counting folders
dir_path = r'C:\Users\Z6\Desktop\3DSEM\Stefan\1.AT\2.June20_v4_final-fr-with-vess2-small-408R\5.ALIGNED\Batch-alignment1\0.source-images\1.large-vess'
#other directory
    #dir_path = r'C:\Users\Z6\Desktop\3DSEM\Stefan\1.AT\2.June20_v4_final-fr-with-vess2-small-408R\5.ALIGNED\Batch-alignment1\0.source-images\2.small-vess'
print(len(os.listdir(dir_path))) #number of files in directory


count = 1 #start at 1 for sake of indexing difference
# Iterate directory
for path in os.listdir(dir_path):
    
    #how to add leading zeroes
    var_counter = f'{count-1:04d}' #need to subtract from 1 from count (number of files) otherwise will iterate incorrectly because indexing at 0 vs. 1
    print(var_counter)

    #make folder for each image
    current_dst_dir = "C:/Users/Z6/Desktop/3DSEM/Stefan/1.AT/2.June20_v4_final-fr-with-vess2-small-408R/5.ALIGNED/Batch-alignment1/0.source-images/5.jupyt-output" + "/" + var_counter
    if not os.path.exists(current_dst_dir):
        os.makedirs(current_dst_dir)
        os.makedirs(current_dst_dir + "/" + "output") # create additional output folder for when doing FIJI image registration
    
    for tiff_file in glob.iglob(os.path.join(source_parent_folder, "**", ('*' + var_counter +'.tif'))): #have to turn off recursive to prevent errors, otherwise detects same image twice and fails
        shutil.copy(tiff_file, current_dst_dir)
    
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        count += 1
print('File count:', count-1) # check file count with counter
```
